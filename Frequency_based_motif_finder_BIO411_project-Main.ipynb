```julia
# opening a file in read_mode
# r is the default mode

function fasta_reader(file_path)
f = open(file_path, "r")

# to count total lines in the file
#line_count = 0			
Sequence_superstring=""
for line in readlines(f)

	# increment line_count
	#global line_count += 1
    if !occursin(r">",line) #Negation of boolean value of whether or not a line has > corresponding to fasta header
	 #println(line_count,' ',line)
     Sequence_superstring*=line
    end

end

# total lines in file
#println("line count is $line_count")

close(f)
#println(Sequence_superstring)
if occursin(r"\n",Sequence_superstring)
    println("Yes there are newline characters we need to remove")
    Sequence_superstring =replace(Sequence_superstring,r"\n"=>"")
end
    return Sequence_superstring
end
```

```julia
function Motif_storer(Sequence_superstring,Motif_length_range)
Storer_Dict = Dict{Int, Dict{Int, String}}()
for Motif_length in Motif_length_range
    Iteration_end =length(Sequence_superstring)-Motif_length+1
    #println(Iteration_end)
    Specific_Length_Dict = Dict{Int, String}()
    for i in 1:Iteration_end
        #println(i,Sequence_superstring[i:i+Motif_length-1])
        Specific_Length_Dict[i] = Sequence_superstring[i:i+Motif_length-1]
        #print(Specific_Length_List)
    end
    Storer_Dict[Motif_length]=Specific_Length_Dict
end
    return Storer_Dict
end
```

```julia
function Motif_Frequency_Calculator(Storer_Dict,Specific_Motif_Length)
    Sample_Test = collect(values(Storer_Dict[Specific_Motif_Length])) #Converting the iterator object to an array
    # Initialize an empty dictionary to store frequencies
    Frequency_Dict = Dict{String,Int}()

    # Calculate frequencies by iterating over all items in the list
    for motif in Sample_Test
        Frequency_Dict[motif] = get(Frequency_Dict, motif, 0) + 1 #Extracting value corresponding to frequency[item] and increasing it
    end

    #println(Frequency_Dict)
    return Frequency_Dict
end
```

```julia
Frequency_Dict = Motif_Frequency_Calculator(Storer_Dict,4)
```

```julia
function Top_n_frequency_storer(Frequency_Dict,Top_n_motif_no)
    # Sort by frequency in descending order and take the top 10
    # Convert dictionary to an array of pairs, sort by frequency in descending order, and take the top 10
    Sorted_Frequency = sort(collect(Frequency_Dict), by = x -> x[2], rev = true)
    Top_n_Frequency_Dict = Dict(Sorted_Frequency[1:min(Top_n_motif_no, length(Sorted_Frequency))]...)  # Ensure at most 10 items

    return Top_n_Frequency_Dict
end
```

```julia
Top_10_frequency = Top_n_frequency_storer(Frequency_Dict,10)
```

```julia
using GLMakie,CairoMakie
function Motif_Frequency_Plotter(Frequency_Dict,Specific_Motif_Length,chosen_backend,chosen_color=:darkred)
    if chosen_backend == "GL"
        GLMakie.activate!()
        xticklabelsize_val = 5/Specific_Motif_Length
    elseif chosen_backend == "Cairo"

        CairoMakie.activate!()
        xticklabelsize_val = 2.5/Specific_Motif_Length
    end

    # Example data (replace with your actual data)
    Motifs = collect(keys(Frequency_Dict))  # Vector{String}
    Frequencies = collect(values(Frequency_Dict))  # Vector{Int64}

    # Map categorical Motifs to integer indices
    Motif_indices = 1:length(Motifs)

    # Create a figure and axis
    fig = Figure(size = (800, 600))
    ax = Axis(fig[1, 1], 
        title = "Frequencies of Motifs of length $Specific_Motif_Length",
        xlabel = "Motifs", 
        ylabel = "Frequencies",
        xticks = (Motif_indices, Motifs),
        xticklabelsize = xticklabelsize_val,
        xticklabelrotation = pi/2,  # Rotate x-tick labels by 90 degrees
        xgridvisible = false,  # Optional, remove the x-grid
        ygridvisible = false)  # Optional, remove the y-grid

    # Plot the bar chart using integer indices
    barplot!(ax, Motif_indices, Frequencies, color=chosen_color)
        
    # Save the figure as a PNG file (or PDF, SVG, etc.)
    if chosen_backend == "Cairo"
         save("Motif_length_$Specific_Motif_Length.svg", fig) #Comment out the save when using GLMakie Backend
    end
    # Display the interactive figure
    return display(fig)
end
```

```julia
function Fasta_Motif_finder(file_path,Motif_length_range=4:20,chosen_backend="Cairo",chosen_color=:darkred,Top_n_motif_no=10)
    """
    The function has arguments as follow:
    Fasta_Motif_finder(file_path,motif_range,chosen_backend)
    file_path: Path to your fasta file
    Motif_length_range: Range of motif lengths given as min_length:max:length, defaults to 4:20
    chosen_backend: Makie backend used to make plot, the values it can take are 
    "GL" for GL and interactive plot and "Cairo" for CairoMakie and non-interactive plot,
    defaults to "Cairo", GL appears not too work to well currently
    chosen_color: Color of the plot, any valid color accepted by Makie of the form :color_name, defaults to dark red
    Top_n_motif_no: Indicates which top (top_n_motif_no) frequency motifs will be saved for later use
    """
    Sequence_superstring = fasta_reader(file_path)
    Storer_Dict=Motif_storer(Sequence_superstring,Motif_length_range)
    # Dictionary to store results for each motif length
    results = Dict()
    Top_motifs = Dict()
    for Specific_Motif_Length in Motif_length_range
        Frequency_Dict = Motif_Frequency_Calculator(Storer_Dict,Specific_Motif_Length)
        plot_result = Motif_Frequency_Plotter(Frequency_Dict,Specific_Motif_Length,chosen_backend,chosen_color)
         # Store the result with the motif length as the key
        Top_n_frequency_dict = Top_n_frequency_storer(Frequency_Dict,Top_n_motif_no)
        results[Specific_Motif_Length] = plot_result
        Top_motifs[Specific_Motif_Length] = Top_n_frequency_dict
    end
    
    return (results,Top_motifs)
end
```
